```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

mu_space = np.linspace(-6, 6, 200)
sigma_space = np.linspace(-6, 6, 200)
mm, ss = np.meshgrid(mu_space, sigma_space)

def neg_log_likelihood(x, mu, logsigma):
    x = x.reshape(-1, 1, 1)
    return np.mean(logsigma + (x - mu) ** 2 / (2 * np.exp(logsigma) ** 2),
                   axis=0)
```

```python
mu = 0.5
sigma = 0.1

rng = np.random.RandomState()
x = rng.randn(1000) * sigma + mu
nll_min = neg_log_likelihood(x, mm, ss).min()
```

```python
mm, ss = np.meshgrid(mu_space, sigma_space)

rng = np.random.RandomState()
x = rng.randn(32) * sigma + mu
nll = neg_log_likelihood(x, mm, ss)
# m = plt.pcolormesh(mm, ss, nll)
plt.contour(mm, ss, nll, levels=nll_min + np.logspace(-3, 5, 15))
plt.xlabel("mu")
plt.ylabel("log sigma");
# plt.colorbar(m);
```
